```python
from chardet.universaldetector import UniversalDetector
import os
from os.path import isfile, join
import PyPDF2
import shutil
```

```python
pdf_dir = "/Users/jeriwieringa/Dissertation/text/corpora-incoming/periodicals/"
staging = "/Users/jeriwieringa/Dissertation/text/text/staging/"
```

## Extract text from each page and save to file.

```python
# %load ../ocr_evaluation_and_correction/corpus_reports/shared_elements/load_corpus.py
corpus = (f for f in os.listdir(pdf_dir) if not f.startswith('.') and isfile(join(pdf_dir, f)))
```

```python
## Following pattern used by NLP Group

for filename in corpus:
    # Open the PDF and load as PyPDF2 Reader object.
    pdf = open('/Users/jeriwieringa/Dissertation/text/corpora-incoming/periodicals/{}'.format(filename),'rb')
    pdfReader = PyPDF2.PdfFileReader(pdf)
    
    # Loop through the pages, extract text and write each page to individual file.
    for page in range(0, pdfReader.numPages):
        pageObj = pdfReader.getPage(page)
        text = pageObj.extractText()
        
        # Compile the page name. Adding one because Python counts from 0.
        pageName = "{}-page{}.txt".format(filename[:-4], page + 1)

        # Write to file
        with open(join(staging, pageName), mode="w", encoding='utf-8') as o:
            o.write(text)
```

## Check Encoding of Text Files

https://www.joelonsoftware.com/2003/10/08/the-absolute-minimum-every-software-developer-absolutely-positively-must-know-about-unicode-and-character-sets-no-excuses/


```python
late_additions = (f for f in os.listdir(staging) if not f.startswith('.') and isfile(join(staging, f)))
```

```python
docs_to_correct = []
```

```python
# Pattern from http://chardet.readthedocs.io/en/latest/usage.html

detector = UniversalDetector()
for filename in late_additions:
    detector.reset()
    for line in open(join(staging, filename), 'rb'):
        detector.feed(line)
        if detector.done: break
    detector.close()
    if not detector.result['encoding'] == 'utf-8':
        print("{}: {}".format(filename, detector.result))
        docs_to_correct.append(filename)
```

## Change Encoding on Files

```python
# Copy files from staging to staging_out

utf8_dir = '/Users/jeriwieringa/Dissertation/text/text/staging_out'

if not os.path.exists(utf8_dir):
    os.makedirs(utf8_dir)

for filename in (f for f in os.listdir(staging) if not f.startswith('.') and isfile(join(staging, f))):
    src_file = join(staging, filename)
    print(src_file)
    shutil.copy(src_file, utf8_dir)
```
