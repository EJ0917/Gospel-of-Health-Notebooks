```python
input_files = ['/Users/jeriwieringa/Dissertation/drafts/data/module-3/2017-04-noun-phrases-2000.txt',
              '/Users/jeriwieringa/Dissertation/archived/experiments/rhSample1920TopicModelTests/RHSamplePhrases.txt'
              ]
```

```python
output_file = "/Users/jeriwieringa/Dissertation/drafts/code/modules/GoH/GoH/preprocess.py"
```

```python
phrases = []
```

```python
for input_file in input_files:
    with open(input_file) as f:
        content = f.read().splitlines()
        print(len(content))
        for phrase in content:
            phrases.append(phrase)
```

```python
unique_phrases = list(set(phrases))
```

```python
noun_phrases = '; '.join(unique_phrases)
```

```python
text="""# -*- coding: utf-8 -*-

ENTITIES = \"""{}\"""

ENTITIES = frozenset(w for w in ENTITIES.split('; ') if w)
""".format(noun_phrases)
```

```python
with open(output_file, 'w') as o:
    o.write(text)
```
